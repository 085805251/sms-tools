# Using an IPython Notebook

This is an example ipython notebook. You can write your observations, code and have results at the sample place, 
connected to each other. This makes it very easy to recall the stuff you've done a long while back (For example,

Let's take an example, and learn to use the ipython notebook in the way. Let's say I want to plot two sinusoids 
together in a single window, along with another sinusoid which is obtained by multiplying them. I want to make an observation
about the figure and keep it for posterity!

1. Create a new cell in the notebook. There are two ways to do this. 
    1. If you are in a cell and want to execute it and move to next cell, you press shift+return keys. It creates a new cell if you are in the last cell of the notebook.
    2. From the menu, choose Insert > Insert a cell above/below.

I do this, and write some code snippet that plots a sine wave and an exponential (don't worry about the details, we'll learn it soon).

## The mind-blowing plot

```python
x = linspace(0, 10, 3000)
y1 = sin(x)
y2 = cos(x)

plot(x, y1)
plot(x, y2)
plot(x, y1*y2)
```

**_Tip:_** Once I'm done coding, I press **shift+return** if I want to see the output and move to next cell. 
Whereas, if I want to just execute but stay in the same cell, then I press **ctrl+return**. Try it now!

The newly created cell will be a 'code' cell by default. If you want to make it a different type of cell (eg: text with some markup, 
headings or raw text), just select it from the menu on the top. Now, to note my ground-breaking observations from this plot, I create
a new cell and make it a 'markdown' cell.

## Publication-worthy observations
